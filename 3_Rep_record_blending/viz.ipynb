```python
import study_ordered_blending
import matplotlib.pyplot as plt
import pandas as pd
```

```python
study_ordered_blending.main()
```

```python
df = pd.pandas.read_csv("agent0-blending-log-2022-03-22.csv")
```

```python
disconnected_df = df.loc[df['network'] == 'disconnected']
fully_df = df.loc[df['network'] == 'fully-connected']
ring_df = df.loc[df['network'] == 'ring']
```

```python
x = disconnected_df['round'].values.tolist()
y = disconnected_df['c_blending_value'].values.tolist()
z = disconnected_df['d_beldning_value'].values.tolist()
plt.plot(x, y,label='disconnected cooperate')
plt.plot(x, z,label='disconnected defect')
plt.legend()
plt.show()
```

```python
x = fully_df['round'].values.tolist()
y = fully_df['c_blending_value'].values.tolist()
z = fully_df['d_beldning_value'].values.tolist()
plt.plot(x, y,label='fully-connected cooperate')
plt.plot(x, z,label='fully-connected defect')
plt.legend()
plt.show()
```

```python
x = ring_df['round'].values.tolist()
y = ring_df['c_blending_value'].values.tolist()
z = ring_df['d_beldning_value'].values.tolist()
plt.plot(x, y,label='ring cooperate')
plt.plot(x, z,label='ring defect')
plt.legend()
plt.show()
```

## Disconnected Network

```python
cooperate_CC = disconnected_df.loc[disconnected_df['cooperate_rep'] == 'CC']
cooperate_CD = disconnected_df.loc[disconnected_df['cooperate_rep'] == 'CD']
cooperate_DC = disconnected_df.loc[disconnected_df['cooperate_rep'] == 'DC']
cooperate_DD = disconnected_df.loc[disconnected_df['cooperate_rep'] == 'DD']
```

```python
x_cc = cooperate_CC['round'].values.tolist()
y_cc = cooperate_CC['c_blending_value'].values.tolist()

x_cd = cooperate_CD['round'].values.tolist()
y_cd = cooperate_CD['c_blending_value'].values.tolist()

x_dc = cooperate_DC['round'].values.tolist()
y_dc = cooperate_DC['c_blending_value'].values.tolist()

x_dd = cooperate_DD['round'].values.tolist()
y_dd = cooperate_DD['c_blending_value'].values.tolist()
```

```python
plt.plot(x_cc, y_cc,label='cooperate_CC')
plt.plot(x_cd, y_cd,label='cooperate_CD')
plt.plot(x_dc, y_dc,label='cooperate_DC')
plt.plot(x_dd, y_dd,label='cooperate_DD')

plt.legend()
plt.title("agent 0 in disconnected network")
plt.show()
```
