```python
df = pd.pandas.read_csv("agent0-activation-log-2022-04-05.csv")
```

```python
disconnected_df = df.loc[df['network'] == 'disconnected']
fully_df = df.loc[df['network'] == 'fully-connected']
ring_df = df.loc[df['network'] == 'ring']
```

```python
def activation_graph(name,df):
    x = df['step'].values.tolist()
    y_1 = df['c_ins_1_activation'].values.tolist() #5
    y_2 = df['c_ins_2_activation'].values.tolist() #0
    y_3 = df['d_ins_1_activation'].values.tolist()
    y_4 = df['d_ins_2_activation'].values.tolist()

    plt.plot(x, y_1,label='cooperate, 5')
    plt.plot(x, y_2,label='cooperate 0')
    plt.plot(x, y_3,label='defect 5')
    plt.plot(x, y_4,label='defect 0')
    plt.title(name+ " activation value")
    plt.legend()
    plt.show()

def retrival_graph(name,df):
    x = df['step'].values.tolist()
    y_1 = df['c_ins_1_retP'].values.tolist() #5
    y_2 = df['c_ins_2_retP'].values.tolist() #0
    y_3 = df['d_ins_1_retP'].values.tolist()
    y_4 = df['d_ins_2_retP'].values.tolist()

    plt.plot(x, y_1,label='cooperate, 5')
    plt.plot(x, y_2,label='cooperate 0')
    plt.plot(x, y_3,label='defect 5')
    plt.plot(x, y_4,label='defect 0')
    plt.title( name + " retrival probability")
    plt.legend()
    plt.show()
```

## Disconnected

```python
activation_graph("disconnected",disconnected_df)
```

```python
retrival_graph("disconnected",disconnected_df)
```

## Ring

```python
activation_graph("ring",ring_df)
```

```python
retrival_graph("ring",ring_df)
```

## Fully connected

```python
activation_graph("fully-connected",fully_df)
```

```python
retrival_graph("fully-connected",fully_df)
```
