### Fully connected

```python
df_fully_connected['1-2-choice'] = df_fully_connected['player 1 choice'].astype(str) + df_fully_connected['player 2 choice'].astype(str)
df_fully_connected_r = df_fully_connected[["participant cluster", "round","1-2-choice"]]
participant_round_counts = df_fully_connected_r.groupby(['1-2-choice','round']).count().reset_index()
#participant_round_counts.rename(colums={'1-2-choice': 'choice of agent A and B'}, inplace=True)

pvt = pandas.pivot_table(participant_round_counts, values='participant cluster', index='round',columns=['1-2-choice'])/300
sns.lineplot(data=pvt).set(title='Fully connected',ylim=(0, 1))
```

### ring

```python
df_ring['1-2-choice'] = df_ring['player 1 choice'].astype(str) + df_ring['player 2 choice'].astype(str)
df_ring_r = df_ring[["participant cluster", "round","1-2-choice"]]

participant_round_counts = df_ring_r.groupby(['1-2-choice','round']).count().reset_index()
#participant_round_counts.rename(colums={'1-2-choice': 'choice of agent A and B'}, inplace=True)

pvt = pandas.pivot_table(participant_round_counts, values='participant cluster', index='round',columns=['1-2-choice'])/300
sns.lineplot(data=pvt).set(title='Ring',ylim=(0, 1))
```

### Disconnected

```python
df_disconnected['1-2-choice'] = df_disconnected['player 1 choice'].astype(str) + df_disconnected['player 2 choice'].astype(str)
df_disconnected_r = df_disconnected[["participant cluster", "round","1-2-choice"]]

participant_round_counts = df_disconnected_r.groupby(['1-2-choice','round']).count().reset_index()
#participant_round_counts.rename(colums={'1-2-choice': 'choice of agent A and B'}, inplace=True)

pvt = pandas.pivot_table(participant_round_counts, values='participant cluster', index='round',columns=['1-2-choice'])/300
sns.lineplot(data=pvt).set(title='Disconnected',ylim=(0, 1))
```
